```python
with open("marvelDCperso.txt", "r") as fic:
    lines = fic.read().split(',')

myList=[]
for line in lines:
    i=line.lower()
    j=i.title()
    myList.append(j)
```

```python
print(myList)
```

```python
textfile = open("persoUppercase1.txt", "w")
for element in myList:
    textfile.write(element + "\n")
textfile.close()
```
